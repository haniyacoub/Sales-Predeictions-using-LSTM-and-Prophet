```
def Predict_Sales(Product_number, Branch_number, Data_Name, Monthly_or_Daily, Per_Branch_or_Per_Product):
    
            import pandas as pd
            import seaborn as sns
            import matplotlib.pyplot as plt
            from fbprophet import Prophet
            from fbprophet.plot import plot_plotly, plot_components_plotly
            name = Data_Name+'.xlsx'
            df = pd.read_excel(name)




            if(Per_Branch_or_Per_Product == 'Per_Branch')&(Monthly_or_Daily == 'Monthly'):

                    df = pd.DataFrame(df[(df['BRANCH_CODE']==Branch_number)&(df['ITEM_CATECORY_CODE']==Product_number)]).groupby('Month_Year').sum()['PRICE_WITH_TAX'].reset_index().rename({'Month_Year':'ds','PRICE_WITH_TAX':'y'},axis=1)
                    m = Prophet(seasonality_mode='multiplicative').fit(df)
                    future = m.make_future_dataframe(periods=1500)
                    fcst = m.predict(future)
                    fig = m.plot(fcst)
                    fcst = m.predict(future)
                    fig = m.plot_components(fcst)
                    future = m.make_future_dataframe(periods=60, freq='MS')
                    fcst = m.predict(future)
                    fig = m.plot(fcst)

            elif(Per_Branch_or_Per_Product == 'Per_Branch')&(Monthly_or_Daily == 'Daily'):

                    df = df = df[(df['BRANCH_CODE']==Branch_number)&(df['ITEM_CATECORY_CODE']==Product_number)].groupby('Date').sum().reset_index()[['Date','PRICE_WITH_TAX']].rename({'Date':'ds','PRICE_WITH_TAX':'y'},axis=1)         
                    m = Prophet()
                    m.fit(df)
                    future = m.make_future_dataframe(periods=365)
                    future.tail()
                    forecast = m.predict(future)
                    forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
                    plt.figure(figsize=(20,10))
                    fig1 = m.plot(forecast)
                    fig2 = m.plot_components(forecast)
                    #plot_plotly(m, forecast)
                    #plot_components_plotly(m, forecast)
                    
            elif(Per_Branch_or_Per_Product == 'Per_Product')&(Monthly_or_Daily == 'Monthly'):

                    df = pd.DataFrame(df[df['ITEM_CATECORY_CODE']==Product_number]).groupby('Month_Year').sum()['PRICE_WITH_TAX'].reset_index().rename({'Month_Year':'ds','PRICE_WITH_TAX':'y'},axis=1)
                    m = Prophet(seasonality_mode='multiplicative', mcmc_samples=300).fit(df)
                    future = m.make_future_dataframe(periods=120, freq='MS')

                    fcst = m.predict(future)
                    fig = m.plot_components(fcst)
                    fig = m.plot(fcst)

            else: #(Per_Branch_or_Per_Product == 'Per_Product')&(Monthly_or_Daily == 'Daily'):

                    df = df[df['ITEM_CATECORY_CODE']==Product_number].groupby('Date').sum().reset_index()[['Date','PRICE_WITH_TAX']].rename({'Date':'ds','PRICE_WITH_TAX':'y'},axis=1)        
                    m = Prophet()
                    m.fit(df)
                    future = m.make_future_dataframe(periods=180)
                    future.tail()
                    forecast = m.predict(future)
                    forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
                    fig1 = m.plot(forecast)
                    fig2 = m.plot_components(forecast)
                    #plot_plotly(m, forecast)
                    #plot_components_plotly(m, forecast)
                    

        
            #plt.figure(figsize=(20,10))
            sns.lineplot(x='ds',y='y',data=df)
```
